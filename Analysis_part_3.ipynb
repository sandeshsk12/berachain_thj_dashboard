```python
import pandas as pd
import ast

df=pd.read_csv('bera_transactions_data.csv')

# Convert string representations to actual lists
df['token_ids'] = df['token_ids'].apply(ast.literal_eval)
df['token_amounts'] = df['token_amounts'].apply(ast.literal_eval)

# Explode 'token_ids' and 'token_amounts'
df_exploded = df.explode(['token_ids', 'token_amounts'])
df_exploded.to_csv('exploded.csv')
```

```python
token_data=pd.read_csv('berachain_token_names.csv',header=None)
token_data=token_data
token_data
```

```python
number_of_tokens=df_exploded[['token_ids','token_amounts']].groupby(by='token_ids').sum()
number_of_tokens=pd.merge(left=number_of_tokens,right=token_data,left_index=True,right_index=True)
number_of_tokens.reset_index(inplace=True)
number_of_tokens.columns=['token_id','token_amounts','NFT']
number_of_tokens.sort_values(by='token_amounts',ascending=False)
```

```python
number_of_tokens.to_csv('token_distribution.csv')
```
