```python
import requests
import time
import random
import pandas as pd

def split_string(input_string, part_length=64):
    # Use list comprehension to split the string into parts of specified length
    return [input_string[i:i+part_length] for i in range(0, len(input_string), part_length)]

master_df=pd.DataFrame([])
start_hex_number = []
end_hex_number = []
count=[]
z=0

for i in range(762066, 2272230, 5000):#792066 2172230
#     print()
    start_hex_number.append(hex(i))
    end_hex_number.append(hex(i + 10000-1))
    count.append(i)
    # if i==792066:
#     break

# Iterate over the pairs of start and end hex numbers
for start, end,count in zip(start_hex_number, end_hex_number,count):
    z=z+1
    print('started ',count,end=' ')
    df=pd.DataFrame([])
    time.sleep(2+round(random.uniform(-1,2),2))
    # Define the URL
    try:
        url = 'https://docs-demo.bera-artio.quiknode.pro/'

        # Define the JSON payload for the eth_getLogs method
        payload = {
            'jsonrpc': '2.0',
            'method': 'eth_getLogs',
            'params': [{
                'fromBlock': start,
                'toBlock': end,
                'address': ['0xd9c0d89a9196d29d7b840e8225e550705cc02aa0'],
                'topics': ['0x4a39dc06d4c0dbc64b70af90fd698a233a518aa5d07e595d983b8c0526c8f7fb']
            }],
            'id': 1
        }

        response = requests.post(url, json=payload)
        response.raise_for_status()  # Raise an exception for HTTP errors
        data = response.json()
    except Exception as e:
        print()
        print(count,':',e)

    log_entries = []

    try:
        for log in data['result']:
            parts = split_string(log['data'][2:])
            number_of_records=int(parts[2],16)
            ids=[int(x, 16) for x in parts[3:number_of_records+3]]
            amounts=[int(x, 16) for x in parts[len(ids)+3+1:]]
            log_entry = {
                    'blockNumber': log['blockNumber'],
                    'transactionHash': log['transactionHash'],
                    'transactionIndex': log['transactionIndex'],
                    'address': log['address'],
                    'user_address': '0x'+log['topics'][1][-40:],
                    'topics0': log['topics'][0] if len(log['topics']) > 0 else None,
                    'topics1': log['topics'][1] if len(log['topics']) > 1 else None,
                    'topics2': log['topics'][2] if len(log['topics']) > 2 else None,
                    'topics3': log['topics'][3] if len(log['topics']) > 3 else None,
                    'log_data': log['data'],
                    'logIndex': log['logIndex'],
                    'removed': log['removed'],
                    'token_ids':ids,
                    'token_amounts':amounts
              }
            log_entries.append(log_entry)

          # Construct the DataFrame from the list of dictionaries
        df = pd.DataFrame(log_entries)
        df.to_csv('small_files/bera_chain_part_{}.csv'.format(count))
#         master_df=pd.concat([master_df,df])
        print('Finished ',count,end=' ')
        print()
    except Exception as e:
        print('Failed ',count, ' :' ,e)
#         print(data)
        print()
        pass
    if z%12==0:
        time.sleep(10+round(random.uniform(-1,2),2))


```
