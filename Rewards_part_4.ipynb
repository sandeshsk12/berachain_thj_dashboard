```python
block_data=pd.read_csv('block_date.csv')
```

```python
rewards=pd.read_csv('rewards.csv')
rewards.reset_index(inplace=True)
rewards.columns=['token_ids','NFT','Reward','Max']
rewards
```

```python
block_data=block_data[['block_number', 'block_number_hex', 'block_timestamp',
       'block_datetime', 'block_date']]
```

```python
txn_data=txn_data[['blockNumber', 'transactionHash',
       'transactionIndex', 'address', 'user_address', 'topics0', 'topics1',
       'topics2', 'topics3', 'log_data', 'logIndex', 'removed', 'token_ids',
       'token_amounts']]
```

```python
# temp = txn_data[txn_data['user_address'].isin(['0xb129114fae2d533629d442657ea5bc1bca5a0364','0x722c8e2b51312a591fe0d20165f544799bf92004', '0xd4b8130df6e6ec6f8061e7dde103e09d25fb58e0'])]
temp=txn_data.copy()
```

```python
temp_merge=pd.merge(temp,block_data,left_on='blockNumber',right_on='block_number_hex')
temp_merge
```

```python
rewards_merged=pd.merge(temp_merge,rewards,left_on='token_ids',right_on='token_ids')
rewards_merged
```

```python
rewards_merged['total_reward']=np.maximum(rewards_merged['Max'],rewards_merged['token_amounts']*rewards_merged['Reward'])
```

```python
rewards_merged
```

```python
user_rewards_date=rewards_merged[['user_address','block_date','total_reward']].groupby(['user_address','block_date'],as_index=False).sum()
# user_rewards_date['cummulative_reward']=user_rewards_date['total_reward'].cumsum()
user_rewards_date
```

```python
rewards_daily_change=user_rewards_date.groupby(['user_address','block_date']).sum()
rewards_daily_change['cumsum_total_reward'] = rewards_daily_change.groupby('user_address')['total_reward'].cumsum()
rewards_daily_change=rewards_daily_change.reset_index()
# rewards_summed.set_index(['user_address','block_date'],inplace=True)
rewards_daily_change
```

```python
df = rewards_daily_change.copy()
df['block_date'] = pd.to_datetime(df['block_date'])

# Creating a date range from 2024-02-26 to 2024-05-18
date_range = pd.date_range(start='2024-02-26', end='2024-05-18')
date_df = pd.DataFrame(date_range, columns=['date'])

# Initialize an empty DataFrame to hold the processed data
rewards_daily = pd.DataFrame()

# Group by 'user_address' and process each group
for _, group in df.groupby('user_address'):
    # Merge with the date range
    merged_group = pd.merge(date_df, group, left_on='date', right_on='block_date', how='left')
    # Forward fill missing data
    merged_group['total_reward'].fillna(method='ffill', inplace=True)
    merged_group['cumsum_total_reward'].fillna(method='ffill', inplace=True)
    # Fill initial missing values with 0
    merged_group.ffill(inplace=True)
    # Append to the final DataFrame
    rewards_daily = pd.concat([rewards_daily, merged_group], ignore_index=True)
    rewards_daily.dropna()
# Display the final DataFrame
rewards_daily['cumsum_total_reward'] = rewards_daily['cumsum_total_reward'].clip(upper=160)

```
